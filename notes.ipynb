What is web framework

Web framework is collection of modules , packages and libraries design to speed up developement


Django follows MVT (Model View Controller)


Model (M): This is the Data Layer. It defines the structure of your database using Python classes. Each model usually maps to a single database table, and Django's Object-Relational Mapping (ORM) allows you to interact with data without writing SQL.View (V): This is the Logic Layer. It acts as a bridge between the Model and the Template. The View receives a web request, fetches the necessary data from the Model, processes it, and decides which Template to use for the response.Template (T): This is the Presentation Layer. It is an HTML file that uses Django Template Language (DTL) to show dynamic data. It defines how the web page should look and where the data provided by the View should be placed.
